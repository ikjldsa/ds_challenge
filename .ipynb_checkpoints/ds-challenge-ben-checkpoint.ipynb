# Deep Learning

* Advantages:
    * Strong adaptatibility on almost any data with any input and outpt.
    * Good Performance even with highly correlated data
    * Will not be distracted by variable with low predictive value.
 
* Disadvantages:
    * Usually require more data to get good performance due to less model assumption.
    * Due to the highly complex neuron layers, it's usally more difficult to trace the relationship on how a particular input varaible impacted the output.  
     

```python
y = 'CLAIM_FLAG'
x = ['HOME_CHILDREN',
    'JOB_TENURE',
    'INCOME',
    'HOME_VAL',
    'MARITAL_STATUS',
    'GENDER',
    'EDUCATION',
    'OCCUPATION',
    'TRAVTIME',
    'CAR_USE',
    'CAR_COST',
    'PTIF',
    'CAR_TYPE',
    'OLDCLAIM_LOG',
    'CLM_FREQ',
    'REVOKED',
    'MVR_PTS',
    'CAR_AGE',
    'CITY_RURAL',
    'AGE_BAND',
    'INCOME_FLAG',
    'HOME_VAL_FLAG',
    'OLDCLAIM_FLAG',
    'CHILD_DRIVE_FLAG',
    'HOME_CHILDREN_FLAG']
```

```python
dl_1 = H2ODeepLearningEstimator(distribution="bernoulli",
                               hidden=[100,100],
                               epochs=100,
                               train_samples_per_iteration=-1,
                               activation="TanhwithDropout",
                               hidden_dropout_ratios=[0.1, 0.1],                               
                               standardize=True,
                               stopping_metric = 'MSE',
                               nfolds=5,
                               fold_assignment='Modulo',
                               keep_cross_validation_predictions = True,
                               model_id='dl_1',
                               seed=123)

dl_1.train(x=x,
           y=y,
           training_frame=train,
           validation_frame=test)     

dl_1_pref = dl_1.model_performance(valid=True)
dl_1_pref
```

```python
dl_1_pref.plot()
```

```python
dl_1.varimp_plot(20)
```

```python
dl_2 = H2ODeepLearningEstimator(distribution="bernoulli",
                               hidden=[100,100],
                               epochs=100,
                               train_samples_per_iteration=-1,
                               activation="RectifierWithDropout",
                               hidden_dropout_ratios=[0.1, 0.1],                               
                               standardize=True,  
                               stopping_metric = 'MSE',
                               nfolds=5,
                               fold_assignment='Modulo',
                               keep_cross_validation_predictions = True,
                               model_id='dl_2',
                               seed=123)
dl_2.train(x=x,
           y=y,
           training_frame=train,
           validation_frame=test)     

dl_2_pref = dl_2.model_performance(valid=True)
```

```python
dl_2.varimp_plot(20) 
```

# Random Forest

Also konwn as Tree bagging, consist of many decision trees, and predicting output by ensembling output from all trees.

* Advantages
    * rebrust and allow highly correlated inputs.
    * work well on classificaiton problems, due to its decision tree nature.
    * Usally easy to fine tune hyper-parameter, 
        * Number of trees - just need to set a high enough number for the results to converge.
        * Tree depths - Usally somewhere arround sqroot(number of paramters)

* Disadvantage
    * Not really good at solving non-linear problem.
    * Does not work well on numeric variable, have to break down to band. i.e. AGE -> AGE_BAND
    * Hard to understand the direct correlation between input and response variable due to random tree nature.  

```python
y = 'CLAIM_FLAG'
x = [
    'JOB_TENURE',
    'INCOME',
    'HOME_VAL',
    'MARITAL_STATUS',
    'GENDER',
    'EDUCATION',
    'OCCUPATION',
    'TRAVTIME',
    'CAR_USE',
    'CAR_COST',
    'PTIF',
    'CAR_TYPE',
    'OLDCLAIM_LOG',
    'CLM_FREQ',
    'REVOKED',
    'MVR_PTS',
    'CAR_AGE',
    'CITY_RURAL',
    'AGE_BAND',
    'INCOME_FLAG',
    'HOME_VAL_FLAG',
    'OLDCLAIM_FLAG',
    'CHILD_DRIVE_FLAG',
    'HOME_CHILDREN_FLAG']
```

```python
rf = H2ORandomForestEstimator(nfolds=5,
                              fold_assignment='Modulo',
                              keep_cross_validation_predictions = True,
                              model_id='rf',
                              ntrees = 100)
rf.train(x=x,
        y=y,
        training_frame=train,
        validation_frame=test)
```

```python
rf.plot()
```

Model performance has converged at ~ 40 - 50.

```python
print("auc",rf.auc(valid=True))
print("mse",rf.mse(valid=True))
```

```python
rf.confusion_matrix(valid=True)
```

```python
rf.varimp_plot(20)
```

# Ensemble Model

* Combine Result form dl_1 and rf

```python
stack = H2OStackedEnsembleEstimator(training_frame=train,
                                    validation_frame=test,
                                    base_models=[dl_1, rf],
                                    seed=123,
                                    model_id='stack'
                                   )
```

```python
stack.train(x=x, y=y, training_frame=train, validation_frame=test)
```

```python
stack.confusion_matrix(valid=True)
```

# Summary

I have created 3 models above, dl_1, dl_2, rf. All 3 produce similar level of performances, with dl_2 perform slightly better if measured by AUC value. <br>

For binary classificaiton problem, the goal is to try to get a balance between maximising precision(True Positive/All Predicted Positve) and recall(True Positive/All Positive). These 2 values varibles vary depend on the threshold you set on the preditect value which is between 0 and 1. <br>

By increasing the threshold fro p-value, we not only get more positive prediction, but also include more false positive because prediction with lower p-value will be classed as a possitive prediction. Therefore a good model should mantain a good TP:FP ratio in most probability threshold values(see plot above). AUC caculate how this ratio changes across different threshold value which is a useful indicator for a good model. <br>

The confusion matrix give s frequency breakdown of how the prediction result looks like when compare with the actual data. The package I used above, default to select the confusion at maximum f1 score, which equal to 2 * (precison * recall)/(precisoin + recall) - harmonic mean. <br>

The stack ensemble algorithm is used to combine multiple models prediction into one, hoping boots the model accuracy.

# Summary

Unlike the first problem on solving claim probability which the response variable [0,1]. Claim cost can range between [0, inf). That's why I changed the distribution paramter on both models, to a distirbution with range [0, inf). <br>

In my opinion, it is generally more diificult to determine how good a predicttion model  is if the response varialbe is continous as a prediction is always going to be different to the actual value. A good measure is to measure the mean squared error, which is the average difference between all prediction and the actual value. <br>

The deep-learning model seems to have performance better than the linear regression with a lower MSE(a measure of culmulative distance between predicted value vs real data). <br>

Deep-learning MSE: 0.6746141500405386 <br>
Regression MSE: 0.6148846922277535 <br>

There are of course many more factor to consider, for example the model may only be working well in a particular range of the response, and start spreading with a bigger variance on higher value claims. <br>

A y vs y_hat will give us a good picture on identifying these problem if we can spot any pattern on graph. (see above)

Now that we have computed our model for "Claim Probability" and "Claim Cost". We can then consider the combine risk of both variable. <br />

In statistics, this risk can be model using something similar to a compound poisson distribution, except for simplicity reason, we have assume the maximum number of claim per policy to 1 claim in the model above. <br />

The combined risk can be model using the following fomulas:

Assume I ~ Probaiblity of a policy making a claim = 1 (if claim), 0 Otherwise, <br>
Assume X ~ Cost of the claim, given I = 1. <br>

Y = I * X

The expected value of the combined risk can be derived from simple statistical method: <br>

E[Y] = E[I] * E[X] <br>
Var[Y] = E[I]Var[X] + (E[x))^2 * Var[I]) <br>

Note: Due to the complex nature of neural network and random forest, many of the model assumptions on mean and variance may not hold.
